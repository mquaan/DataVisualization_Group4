```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('data/world_data.csv')
df
```

```python
df.dropna(subset=['Country'], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
#filling missing value by backfill
df.fillna(method='bfill', inplace=True)
#filling missing value by forwardfill
df.fillna(method='ffill', inplace=True)
df
```

```python
df['Year'] = df['Year'].astype(int)
df['Population'] = df['Population'].astype(int)
```

```python
import requests
BASE_URL = 'http://api.worldbank.org/v2/'
all_countries_url = f'{BASE_URL}countries?format=json&per_page=300'
response = requests.get(all_countries_url)

if response.status_code == 200:
    countries_list = [country['id'] for country in response.json()[1]]
    name_list = [country['name'] for country in response.json()[1]]

    countries_df = pd.DataFrame({'Country': name_list, 'Country Code': countries_list})

merged_df = pd.merge(df, countries_df, on='Country', how='left')
merged_df
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
merged_df = pd.merge(merged_df, world[['iso_a3', 'continent']], left_on='Country Code', right_on='iso_a3', how='left')

merged_df.drop(columns=['iso_a3'], inplace=True)
merged_df.rename(columns={'continent': 'Continent'}, inplace=True)
merged_df
```

```python
countries_with_no_continent = merged_df[merged_df['Continent'].isnull()]['Country']
merged_df.drop(merged_df[merged_df['Country'].isin(countries_with_no_continent)].index, inplace=True)
merged_df.reset_index(drop=True, inplace=True)
merged_df
```

```python
def get_status(gdp):
    if gdp >= 22000:
        return 'Developed'
    elif 1025 < gdp < 22000:
        return 'Developing'
    else:
        return 'Least Developed'

merged_df_2022 = merged_df[merged_df['Year'] == 2022].copy()
merged_df_2022['Status'] = (merged_df_2022['GDP'] / merged_df_2022['Population']).apply(get_status)

merged_df = pd.merge(merged_df, merged_df_2022[['Country', 'Status']], on='Country', how='left')
merged_df
```

```python
merged_df.to_csv('data/world_data_cleaned.csv', index=False)
```
